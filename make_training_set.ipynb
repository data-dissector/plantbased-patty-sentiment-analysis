```python
import pandas as pd
```

### Import NLTK-VADER to create training sets of tweets labeled with sentiment (-1 to 1)

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

## Impossible Burger

### Read in tweets, excluding 1/4 of them to save for test set

```python
df = pd.read_csv('impossible_text_training_tweets.csv', nrows=30000)
```

```python
df
```

```python
negative = []
neutral = []
positive = []
compound = []

sid = SentimentIntensityAnalyzer()
```

```python
for i in df.index:
    scores = sid.polarity_scores(df['tweet'].iloc[i])
    negative.append(scores['neg'])
    neutral.append(scores['neu'])
    positive.append(scores['pos'])
    compound.append(scores['compound'])
```

```python
df['negative'] = negative
df['neutral'] = neutral
df['positive'] = positive
df['compound'] = compound
```

```python
df.index
```

### Training set of tweets labeled with sentiment (-1 to 1)

```python
df
```

```python
df.to_csv('final_impossible_text_training_tweets.csv', index=False)
```
