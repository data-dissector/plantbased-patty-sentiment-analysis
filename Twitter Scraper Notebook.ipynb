```python
import os
```

```python
import pandas as pd
```

```python
tweet_count = 40000
text_query = "impossible burger"
since_date = "2012-01-01"
until_date = "2021-05-09"
```

```python
os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> training_tweets.json'.format(tweet_count, since_date, text_query, until_date))
```

```python
tweets_df = pd.read_json('training_tweets.json', lines=True)
```

```python
tweets_df.to_csv('training_tweets.csv', sep=',', index=False)
```
