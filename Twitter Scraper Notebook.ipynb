### Import pandas to convert file to dataframe

```python
import pandas as pd
```

## Scrape Tweets by Text Search

### Impossible Burger

```python
tweet_count = 40000
text_query = "impossible burger"
since_date = "2012-01-01"
until_date = "2021-09-09"
```

```python
os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> impossible_text_training_tweets.json'.format(tweet_count, since_date, text_query, until_date))
```

```python
tweets_df = pd.read_json('impossible_text_training_tweets.json', lines=True)
```

```python
tweets_df
```

```python
tweets_df.to_csv('impossible_text_training_tweets.csv', sep=',', index=False)
```
